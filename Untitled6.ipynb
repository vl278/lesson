```python
# будем делать сводную таблицу
user_movies_matrix=ratings.pivot(index='movieId',columns='userId',values='rating')
user_movies_matrix
```

```python
user_movies_matrix.fillna(0,inplace=True)
user_movies_matrix
```

```python
# удалить неактивных пользователей и мало оцененных фильмов
user_rates=ratings.groupby('userId')['rating'].agg('count')
user_rates
# оставляем пользователе с более 50 оценок
```

```python
movies_rates=ratings.groupby('movieId')['rating'].agg('count')
movies_rates
# филбмы у которых оценок > 10 оценок
```

```python
user_filter=user_rates[user_rates>50].index
movies_filter=movies_rates[movies_rates>10].index
```

```python
user_movies_matrix = user_movies_matrix.loc[:,user_filter]
```
